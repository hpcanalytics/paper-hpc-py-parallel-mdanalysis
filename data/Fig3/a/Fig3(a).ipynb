```python
import sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use('ggplot')
sns.set_style('ticks')
import matplotlib.ticker as ticker
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
# Enable inline plotting  
%matplotlib inline
```

# Fig 4(a): Comparison on Speed-Up (IO vs No-IO)

```python
data1 = 'df7-comm-updated-rmsd.csv'
data2 = 'df7-comm-updated-rmsd-no-IO.csv'
```

```python
df7_1 = pd.read_csv(data1)
df7_2 = pd.read_csv(data2)
```

```python
import matplotlib as mpl

matplotlib.rcParams['legend.fontsize'] = 9
plt.style.use('ggplot')
sns.set_style('ticks')

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1,1,1)

df = [df7_1, df7_2]
labels = ['With IO','Without IO']

color = sns.color_palette()[0]

df[0].plot(y='speed_up', x='blocks', yerr='std_S', ax=ax, marker='o', 
           color=color, linestyle="-",
           lw=2, label='with I/O')
df[1].plot(y='speed_up', x='blocks', yerr='std_S', ax=ax, marker='o', 
           color=color, linestyle="--",
           lw=2, label='without I/O')
df[1].plot(y='Ideal_Scaling', x='blocks', ax=ax, linestyle='--', color='k', lw=2, label='')

ax.yaxis.grid(which='major',linestyle=':', linewidth='0.5', color='black')
ax.xaxis.grid(which='major',linestyle=':', linewidth='0.5', color='black')
ax.yaxis.grid(which='minor',linestyle=':', linewidth='0.5', color='black')
ax.xaxis.grid(which='minor',linestyle=':', linewidth='0.5', color='black') 

ax.set_xlabel('$N_{Processes}$')
ax.set_ylabel('Speed Up')
ax.set_xlim(1, 72)
ax.set_ylim(0, 72)
sns.despine(offset=10, ax=ax)
lgd = ax.legend(loc='upper center', bbox_to_anchor=(0.50, -0.23),
                handlelength=3,
                frameon=False, fancybox=True, shadow=True, ncol=2)
plt.tight_layout()
fig_name = 'Comparison_Speed_UP_with_and_without_IO'
fig.savefig('{}.pdf'.format(fig_name),bbox_extra_artists=(lgd,), bbox_inches='tight')
```
